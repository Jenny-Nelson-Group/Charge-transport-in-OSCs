```python
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
def Calc_CT(E1,E2,J):
    H=np.zeros((2,2))
    H[0,0]=E1
    H[1,1]=E2
    H[0,1]=J
    H[1,0]=J
    evals,evecs=np.linalg.eig(H)
    eps1=evals[0]
    eps2=evals[1]
    print eps1,eps2
    occs1=evecs[:,0]*evecs[:,0]
    occs2=evecs[:,1]*evecs[:,1]
    print occs1,occs2
    return eps1,eps2,occs1,occs2
```

```python
def Plot_Levels(E1,E2,J,Title):
    fig=pl.figure()
    eps1,eps2,occs1,occs2=Calc_CT(E1,E2,J)
    pl.plot([0.25,0.45],[E1,E1],color="b")
    pl.plot([0.75,0.95],[E2,E2],color="r")
    
    pl.plot([0.52,0.52+(0.16*occs1[0])],[eps1,eps1],color="b")
    pl.plot([0.52+(0.16*occs1[0]),0.68],[eps1,eps1],color="r")
    
    pl.plot([0.52,0.52+(0.16*occs2[0])],[eps2,eps2],color="b")
    pl.plot([0.52+(0.16*occs2[0]),0.68],[eps2,eps2],color="r")
    
    
    pl.plot([0.45,0.52],[E1,eps1],linestyle='--',color="gray")
    pl.plot([0.45,0.52],[E1,eps2],linestyle='--',color="gray")
    pl.plot([0.75,0.68],[E2,eps1],linestyle='--',color="gray")
    pl.plot([0.75,0.68],[E2,eps2],linestyle='--',color="gray")
    pl.ylim(-6.8,-2.6)
    pl.title('%s'%(Title))
    pl.show()
    #fig.savefig('%s.png'%(Title))
```
